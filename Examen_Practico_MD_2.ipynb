# Preprocesamiento de datos

```python
# elimar la columana id_estudiante
df = df.drop(columns="id_estudiante")
```

```python
# se remplzan valores faltantes por 0
df["horas_trabajo_semanal"] = df["horas_trabajo_semanal"].fillna(0)
```

```python
#rellenar los valores faltantes usando mediana
from sklearn.impute import SimpleImputer

imputer_num = SimpleImputer(strategy="median")

numericas = [
    "promedio_ciencias",
    "porcentaje_asistencia",
    "entrega_tareas",
    "horas_estudio_semanal",
    "distancia_colegio_km",
    "nivel_autoestima"
]

df[numericas] = imputer_num.fit_transform(df[numericas])
```

```python
#rellenamos valores faltantes usando la moda
imputer_cat = SimpleImputer(strategy="most_frequent")

categoricas = [
    "tendencia_notas",
    "nivel_socioeconomico",
    "nivel_educativo_padres",
    "relacion_con_companeros"
]

df[categoricas] = imputer_cat.fit_transform(df[categoricas])
```

```python
#se comprueba que no existan nulos
df.isnull().sum()
```

```python
#codifica variables ordinales
df["tendencia_notas"] = df["tendencia_notas"].map({
    "Empeorando":0,
    "Estable":1,
    "Mejorando":2
})
```

```python
#codifica variables ordinales
df["participacion_clase"] = df["participacion_clase"].map({
    "Baja":0,
    "Media":1,
    "Alta":2
})
```

```python
#codifica variables ordinales
df["nivel_socioeconomico"] = df["nivel_socioeconomico"].map({
    "Bajo":0,
    "Medio-bajo":1,
    "Medio":2,
    "Alto":3
})
```

```python
#codifica variables ordinales
df["nivel_educativo_padres"] = df["nivel_educativo_padres"].map({
    "Sin estudios":0,
    "Primaria":1,
    "Secundaria":2,
    "Superior":3
})
```

```python
#codifica variables nominales
df = pd.get_dummies(
    df,
    columns=[
        "genero",
        "estructura_familiar",
        "relacion_con_companeros"
    ],
    drop_first=True
)
```

```python
# convertimos la variable categórica (texto) en una variable numérica
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["target"] = encoder.fit_transform(df["target"])
```

```python
#muestra la correspondencia de cada categoria
dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
```

```python
# convierte todos lo boleanos a binario
bool_cols = df.select_dtypes(include="bool").columns

df[bool_cols] = df[bool_cols].astype(int)
```

# Division de datos

```python
#separar variables
X = df.drop("target", axis=1)
Y = df["target"]
```

```python
#verifica dimensiones de las variables
X.shape, Y.shape
```

```python
# divide datos
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.20,
    random_state=42,
    stratify=Y
)
```

```python
#verifica las dimensiones
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# MODELO ML

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced"
)
```

```python
from sklearn.svm import SVC

model_svm = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale",
    class_weight="balanced"
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced"
)
```

# Entrenamiento

```python
model_dt.fit(X_train, Y_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_svm.fit(X_train_scaled, Y_train)
```

```python
model_rf.fit(X_train, Y_train)
```

# Resultados

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Y_pred = model_dt.predict(X_test)

print("Accuracy:", accuracy_score(Y_test,Y_pred))
print("\nReportes:\n", classification_report(Y_test, Y_pred))
print("\nMatriz de confusion:\n", confusion_matrix(Y_test, Y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Y_pred_svm = model_svm.predict(X_test_scaled)

print("Accuracy:", accuracy_score(Y_test, Y_pred_svm))
print("\nReporte:\n", classification_report(Y_test, Y_pred_svm))
print("\nMatriz de confusión:\n", confusion_matrix(Y_test, Y_pred_svm))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Y_pred_train = model_dt.predict(X_train)

print("Accuracy Train:", accuracy_score(Y_train, Y_pred_train))
print("\nReporte Train:\n", classification_report(Y_train, Y_pred_train))
print("\nMatriz de confusión Train:\n", confusion_matrix(Y_train, Y_pred_train))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Y_pred_rf = model_rf.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, Y_pred_rf))
print("\nReporte:\n", classification_report(Y_test, Y_pred_rf))
print("\nMatriz de confusión:\n", confusion_matrix(Y_test, Y_pred_rf))
```
